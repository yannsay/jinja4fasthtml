````
#| export
def export_jinja_template(cell:str, templates_folder:str ="templates"):
    "Export single cell"
    first_line = cell.split('\n')[0]
    file_name = first_line.split("exportjinja")[-1].strip()
    
    if len(cell.split('"""')) == 3:
        content = cell.split('"""')[1]

    elif len(cell.split("'''")) == 3:
        content = cell.split("'''")[1]

    else:
        print("cannot find")
        return None
    
    with open(f"{templates_folder}/{file_name}", "w") as f:
        f.write(content)
````

You have a couple of options to fix this:

1. **Add `#| skip_test` to each individual cell** that uses `find_msgs` (not just at the top of the notebook)
2. **Add `#| eval: false`** to cells that shouldn't be executed during testing
3. **Wrap the code in a try/except** to handle the missing variable gracefully
